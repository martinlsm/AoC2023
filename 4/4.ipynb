```python
# Import stuff and get input
import re
import numpy as np

with open('input', 'r') as f:
    inp = f.readlines()
```

```python
# Solve part 1
res = 0

for line in inp:
    nums = re.findall(r'\d+', line)[1:]
    winning = set((int(x) for x in nums[:10]))
    card_nums = [int(x) for x in nums[10:]]

    matches = 0
    for n in card_nums:
        if n in winning:
            if matches == 0:
                matches = 1
            else:
                matches *= 2
    res += matches

print(res)
```

```python
# Solve part 2
res = 0

matches = {}
for i,line in enumerate(inp):
    nums = re.findall(r'\d+', line)[1:]
    winning = set((int(x) for x in nums[:10]))
    card_nums = [int(x) for x in nums[10:]]

    num_matches = 0
    for n in card_nums:
        if n in winning:
            num_matches += 1
    matches[i] = num_matches

cards = {x:1 for x in range(len(inp))}
for i in range(len(inp)):
    for j in range(i + 1, min(i + 1 + matches[i], len(matches))):
        cards[j] += cards[i]

print(sum(cards.values()))
```
