```python
# Import stuff and get input
import re
import numpy as np

with open('input', 'r') as f:
    inp = f.readlines()

inp = [line.strip() for line in inp]
H = len(inp)
W = len(inp[0])
```

```python
# Solve part 1
res = 0

r,c = 0, 0
parse_num = ''
neighbor_symbol = False
while r != H:
    ch = inp[r][c]
    is_num = re.match(r'\d', ch)

    if is_num:
        # Append to number for parsing.
        parse_num += ch
        
        # Check for neighboring symbols.
        for rr in range(max(0, r - 1), min(H, r + 2)):
            for cc in range(max(0, c - 1), min(W, c + 2)):
                n_ch = inp[rr][cc]
                if not re.match(r'[\d\.]', n_ch):
                    neighbor_symbol = True
    
    if not is_num or c == W - 1:
        # Parse the number previously found, if there is one.
        if parse_num != '' and neighbor_symbol:
            res += int(parse_num)

        # Reset the parsing.
        parse_num = ''
        neighbor_symbol = False

    if c == W - 1:
        c = 0
        r += 1
    else:
        c += 1

print(res)
```

```python
# Solve part 2

# Fill this with lists of part numbers for each gear pos.
part_neighbors = {}

r,c = 0, 0
parse_num = ''
# Fill this with each neighboring gear for a given part number.
gear_neighbors = [] 

while r != H:
    ch = inp[r][c]
    is_num = re.match(r'\d', ch)

    if is_num:
        # Append to number for parsing.
        parse_num += ch
        
        # Check for neighboring gears.
        for rr in range(max(0, r - 1), min(H, r + 2)):
            for cc in range(max(0, c - 1), min(W, c + 2)):
                n_ch = inp[rr][cc]
                if n_ch == '*':
                    if (rr,cc) not in gear_neighbors:
                        gear_neighbors.append((rr,cc))
    
    if not is_num or c == W - 1:
        # Parse the number previously found, if there is one.
        if parse_num != '':
            for p in gear_neighbors:
                if p not in part_neighbors:
                    part_neighbors[p] = []
                part_neighbors[p].append(int(parse_num))

        # Reset the parsing.
        parse_num = ''
        gear_neighbors = []

    if c == W - 1:
        c = 0
        r += 1
    else:
        c += 1

filtered = [parts for parts in part_neighbors.values() if len(parts) == 2]
gear_ratios = [np.prod(ds) for ds in filtered]
print(sum(gear_ratios))
```
