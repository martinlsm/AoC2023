```python
# Import stuff
import re
import numpy as np
```

```python
# Get input
with open('input', 'r') as f:
    inp = [line.strip().split() for line in f.readlines()]
inp = [(x[0], [int(n) for n in x[1].split(',')]) for x in inp]
```

```python

# Solve 1

def solve(pattern, groups, curr_grp_sz=0, indent=''):
    if len(pattern) == 0:
        if len(groups) == 0:
            return 1
        if len(groups) == 1 and groups[0] == curr_grp_sz:
            return 1
        else:
            return 0
    
    if pattern[0] == '#':
        if len(groups) == 0:
            return 0
        if curr_grp_sz >= groups[0]:
            return 0

        res = solve(pattern[1:], groups, curr_grp_sz + 1, indent + ' ')

    elif pattern[0] == '.':
        if curr_grp_sz > 0:
            if curr_grp_sz != groups[0]:
                return 0
            res = solve(pattern[1:], groups[1:], 0, indent + ' ')
        else:
            res = solve(pattern[1:], groups, 0, indent + ' ')
    else:
        assert pattern[0] == '?'
        res = solve('#' + pattern[1:], groups, curr_grp_sz, indent + ' ') + solve('.' + pattern[1:], groups, curr_grp_sz, indent + ' ')

    return res

res = 0
for pattern, groups in inp:
   s = solve(pattern, groups)
   res += s

print(res)
```

```python
# Solve 2

def key(pattern, groups, curr_grp_sz):
    return pattern + ',' + ''.join([str(x) for x in groups]) + ',' + str(curr_grp_sz)


def solve(pattern, groups, cache, curr_grp_sz=0, indent=''):
    if key(pattern, groups, curr_grp_sz) in cache:
        return cache[key(pattern, groups, curr_grp_sz)]

    if len(pattern) == 0:
        if len(groups) == 0:
            return 1
        if len(groups) == 1 and groups[0] == curr_grp_sz:
            return 1
        else:
            return 0
    
    if pattern[0] == '#':
        if len(groups) == 0:
            return 0
        if curr_grp_sz >= groups[0]:
            return 0

        res = solve(pattern[1:], groups, cache, curr_grp_sz + 1, indent + ' ')

    elif pattern[0] == '.':
        if curr_grp_sz > 0:
            if curr_grp_sz != groups[0]:
                return 0
            res = solve(pattern[1:], groups[1:], cache, 0, indent + ' ')
        else:
            res = solve(pattern[1:], groups, cache, 0, indent + ' ')
    else:
        assert pattern[0] == '?'
        res = solve('#' + pattern[1:], groups, cache, curr_grp_sz, indent + ' ') + solve('.' + pattern[1:], groups, cache, curr_grp_sz, indent + ' ')

    cache[key(pattern, groups, curr_grp_sz)] = res
    return res

res = 0
cache = {}
for pattern, groups in inp:
    pattern = '?'.join([pattern] * 5)
    groups = groups * 5
    cache = {}
    s = solve(pattern, groups, cache)
    res += s

print(res)
```
